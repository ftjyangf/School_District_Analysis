```python
import pandas as pd
```

```python
file_to_readst = "Resources/students_complete.csv"

student_df = pd.read_csv(file_to_readst)
student_df.head()

```

```python
st_name_list = student_df.student_name.tolist()
st_name_list
```

```python
st_to_fix = []

for each in st_name_list:
    if len(each.split()) > 2:
        st_to_fix.append(each) 
        
print(st_to_fix)
```

```python
prefixes = []
for name in st_to_fix:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])
prefixes

suffix = []
for suf in st_to_fix:
    if len(suf.split()[-1]) <=3:
        suffix.append(suf.split()[-1])
set(suffix)
```

```python
prefixes_suffixes =["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

for each in prefixes_suffixes:
    student_df.student_name = student_df.student_name.str.replace(each,"")
    print(student_df.student_name)
    
for each in student_df.student_name:
    if len(each.split())>2:
        print(each)
```
