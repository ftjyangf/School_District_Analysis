```python
# join the student table with school table
schoolsMerge_df = pd.merge(students_df,schools_df,on=["school_name","school_name"],how = 'left')
schoolsMerge_df
```

```python
# find mean of reading and math score
schoolsMerge_df['reading_score'].mean()
schoolsMerge_df['math_score'].mean()

# filter the students with math_score greater and equal 70 and number of students
mathpassed = schoolsMerge_df["student_name"][schoolsMerge_df['math_score'] >= 70].count()

# get the number of students and the mean of students who passed the exam
total = schoolsMerge_df["student_name"].count()
mathpassrate = mathpassed/total

readpassed = schoolsMerge_df["student_name"][schoolsMerge_df["reading_score"] >=70].count()

readpassrate = readpassed/total
bothpassed = schoolsMerge_df["student_name"][(schoolsMerge_df['math_score'] >= 70) & (schoolsMerge_df["reading_score"] >=70)].count()

bothrate = bothpassed/total



totalstudent = schoolsMerge_df["student_name"].count()
totalstudent
#total budget

totalbudget = 0
# set only get the unique count
totalschool = set(schoolsMerge_df["budget"])
for r in totalschool:
    totalbudget = r + totalbudget
totalbudget

#average reading
readingaver = schoolsMerge_df["reading_score"].mean()
readingaver
#math average
mathaver = schoolsMerge_df["math_score"].mean()
#total school 
totalschool = len(set(schoolsMerge_df["school_name"]))

```

```python
district_summary_df = pd.DataFrame(
        [{"Total Schools": totalschool,
         "Total Students": totalstudent,
          "Total Budget" : totalbudget,
         "Average Math Score": mathaver,
          "Average Reading Score": readingaver,
          "% Passing Math": mathpassrate * 100,
         "% Passing Reading": readpassrate * 100,
        "% Overall Passing": bothrate * 100}])

district_summary_df['Total Students'] = district_summary_df['Total Students'].map("{:,}".format)

district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map("{:,.2f}".format)

district_summary_df['Average Math Score'] = district_summary_df['Average Math Score'].map("{:.1f}".format)

district_summary_df['Average Reading Score'] = district_summary_df['Average Reading Score'].map("{:.1f}".format)

district_summary_df['% Passing Math'] = district_summary_df['% Passing Math'].map("{:.0f}".format)

district_summary_df['% Passing Reading'] = district_summary_df['% Passing Reading'].map("{:.0f}".format)

district_summary_df['% Overall Passing'] = district_summary_df['% Overall Passing'].map("{:.0f}".format)


district_summary_df.index = ["district_school"]
district_summary_df
```
